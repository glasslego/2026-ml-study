# Colab + GitHub 연동 설정

Private repo를 Colab에서 clone하고, 작업 후 push까지 하는 설정입니다.

## 사전 준비

GitHub **Personal Access Token (PAT)** 이 필요합니다:
1. GitHub → Settings → Developer settings → Personal access tokens → **Fine-grained tokens**
2. **Generate new token** 클릭
3. 설정:
   - Token name: `colab-2026-ml-study`
   - Expiration: 원하는 기간
   - Repository access: **Only select repositories** → `2026-ml-study` 선택
   - Permissions → Repository permissions → **Contents**: Read and write
4. 생성된 토큰을 복사해두세요

---
## 1. GitHub 인증 (매 세션마다 실행)

```python
from google.colab import userdata

# Colab Secrets에 GITHUB_TOKEN 이름으로 PAT를 저장해두면
# 매번 입력할 필요 없이 자동으로 불러옵니다.
# 설정: 좌측 사이드바 🔑 아이콘 → GITHUB_TOKEN 추가
try:
    GITHUB_TOKEN = userdata.get('GITHUB_TOKEN')
    print('Colab Secrets에서 토큰을 불러왔습니다.')
except userdata.SecretNotFoundError:
    GITHUB_TOKEN = input('GitHub PAT를 입력하세요: ')

GITHUB_USER = 'glasslego'
REPO_NAME = '2026-ml-study'
```

## 2. Repo Clone

```python
import os

REPO_URL = f'https://{GITHUB_USER}:{GITHUB_TOKEN}@github.com/{GITHUB_USER}/{REPO_NAME}.git'
WORK_DIR = f'/content/{REPO_NAME}'

if not os.path.exists(WORK_DIR):
    !git clone {REPO_URL} {WORK_DIR}
else:
    print(f'{WORK_DIR} 이미 존재합니다. pull로 최신화합니다.')
    !cd {WORK_DIR} && git pull

os.chdir(WORK_DIR)
print(f'\n현재 디렉토리: {os.getcwd()}')
!git log --oneline -3
```

## 3. Git 사용자 설정

```python
!git config user.name "glasslego"
!git config user.email "your-email@example.com"  # 본인 이메일로 변경하세요
```
